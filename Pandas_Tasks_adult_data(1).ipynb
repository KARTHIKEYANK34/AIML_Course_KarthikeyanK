[link text](https://)Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
bins = [0, 30, 60, Adult['age'].max()]
labels = ['Youth', 'Adult', 'Senior']
Adult['age_group'] = pd.cut(Adult['age'], bins=bins, labels=labels, right=False)
display(Adult.head())
```

Rename the 'educational-num' column to 'education_num'.

```python
Adult = Adult.rename(columns={'educational-num': 'education_num'})
display(Adult.head())
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
Adult['workclass'] = Adult['workclass'].replace('?', 'Unknown')
Adult['occupation'] = Adult['occupation'].replace('?', 'Unknown')
display(Adult.head())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = Adult.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = Adult.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

## Visualize income distribution

### Subtask:
Create a count plot to visualize the distribution of income levels (<=50K and >50K) in the dataset.


**Reasoning**:
To visualize the distribution of income levels, I will create a count plot using seaborn.countplot, setting appropriate labels and title, and then display the plot.



```python
plt.figure(figsize=(8, 6))
sns.countplot(data=Adult, x='income')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

## Visualize income distribution by age group

### Subtask:
Create a count plot to visualize the distribution of income levels (<=50K and >50K) across different age groups ('Youth', 'Adult', 'Senior').


**Reasoning**:
To visualize the income distribution across different age groups, I will create a count plot using seaborn, mapping 'age_group' to the x-axis and 'income' to the hue, and then set the figure size, title, and axis labels for better readability.



```python
plt.figure(figsize=(10, 7))
sns.countplot(data=Adult, x='age_group', hue='income')
plt.title('Income Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

## Visualize marital status distribution

### Subtask:
Create a bar plot to visualize the distribution of marital statuses in the dataset.


**Reasoning**:
To visualize the distribution of marital statuses, I will create a bar plot using the `marital_status_counts` Series, setting the figure size, title, and axis labels, and rotating x-axis labels for readability.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x=marital_status_counts.index, y=marital_status_counts.values)
plt.title('Distribution of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize the distribution of hours-per-week

### Subtask:
Create a histogram to visualize the distribution of the 'hours-per-week' column.

#### Instructions
1. Set the figure size to 10x6 inches using `plt.figure()`.
2. Create a histogram using `seaborn.histplot()` with 'hours-per-week' on the x-axis.

**Reasoning**:

**Reasoning**:
To visualize the average capital-gain by education level, I will first calculate the average 'capital-gain' grouped by 'education' and then create a bar plot using seaborn, setting appropriate labels, title, and figure size, and rotating x-axis labels for readability.



```python
average_capital_gain_by_education = Adult.groupby('education')['capital-gain'].mean()

plt.figure(figsize=(12, 7))
sns.barplot(x=average_capital_gain_by_education.index, y=average_capital_gain_by_education.values)
plt.title('Average Capital Gain by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Capital Gain')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
